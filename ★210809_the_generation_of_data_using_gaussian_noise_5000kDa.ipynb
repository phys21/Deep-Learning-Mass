# The generation of frequency data using gaussian noise

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import cv2
```

making noise

```python
noise = np.random.normal(0,15771.278,2000) #평균, 분산, 갯수
#noise_range = noise*57700 
#sns.distplot(noise)
plt.hist(noise, bins=100)#구간을 몇개로 나눌 것인지
plt.show()
```

confirming noise distribution

```python
sns.distplot(noise)
plt.hist(noise, density=True, bins=100)
plt.show()
```

the algorithm to generate resonance frequency data

```python
#1줄 데이터(10000개의 포인트)
data0 = []
for i in range(0, 2000):
    a = 26567336.36
    data0.append(a)
    b = np.transpose(data0)
print(b)
```

```python
#10000개의 포인트가 3000개 있는 데이터셋
data1 = []
for i in range(0, 5500):
    data1.append(b)
print(data1[0])
```

Injecting noise into the initial resonance frequency

```python
data2 = []
for i in range(0, 5500):
    c = np.random.normal(0,15771.278,2000)+data1[i]
    data2.append(c)
print(data2)
```

Checking the noise injected into the resonance frequency data

```python
plt.plot(data2[0])
plt.show()
```

Save the frequency data including noise in excel format

```python
final_rf = pd.DataFrame(data2)
final_rf.to_csv('210809_initial_freq_with_noise_2000point_5000kDa.csv')
```

Injecting noise into the frequency data accoring to mass change

```python
data3 = []
for i in range(0, 2000):
    a = 26567336.36
    data3.append(a)
    b = np.transpose(data3)
print(b) 

data3_1 = []
for i in range(0,5500):
    data3_1.append(b)
print(data3_1[-1])
```

```python
data4 = []
for i in range(0, 5500):
    c = np.random.normal(0,15771.278,2000)+data3_1[i]
    data4.append(c)
print(data4[0])
```

```python
d4 = pd.DataFrame(data4)
```

```python
#13210 is the frequency value difference for 1000kDa.
mass1 = []
for i in range(0, 500):
    a = d4[i:i+1]
    mass1.append(a)
print(mass1[-1])

mass2 = []
for i in range(500, 1000):
    a = d4[i:i+1] - 13210/2
    mass2.append(a)
print(mass2[0])

mass3 = []
for i in range(1000, 1500):
    a = d4[i:i+1] - 13210
    mass3.append(a)
print(mass3[0])

mass4 = []
for i in range(1500, 2000):
    a = d4[i:i+1] - 13210*3/2
    mass4.append(a)
print(mass4[0])

mass5 = []
for i in range(2000, 2500):
    a = d4[i:i+1] - 13210*2
    mass5.append(a)
print(mass5[0])

mass6 = []
for i in range(2500, 3000):
    a = d4[i:i+1] - 13210*5/2
    mass6.append(a)
print(mass6[0])

mass7 = []
for i in range(3000, 3500):
    a = d4[i:i+1] - 13210*3
    mass7.append(a)
print(mass7[-1])

mass8 = []
for i in range(3500, 4000):
    a = d4[i:i+1] - 13210*7/2
    mass8.append(a)
print(mass8[0])

mass9 = []
for i in range(4000, 4500):
    a = d4[i:i+1] - 13210*4
    mass9.append(a)
print(mass9[0])

mass10 = []
for i in range(4500, 5000):
    a = d4[i:i+1] - 13210*9/2
    mass10.append(a)
print(mass10[0])

mass11 = []
for i in range(5000, 5500):
    a = d4[i:i+1] - 13210*5
    mass11.append(a)
print(mass11[0])
```

```python
m1 = np.array(mass1)
m2 = np.array(mass2)
m3 = np.array(mass3)
m4 = np.array(mass4)
m5 = np.array(mass5)
m6 = np.array(mass6)
m7 = np.array(mass7)
m8 = np.array(mass8)
m9 = np.array(mass9)
m10 = np.array(mass10)
m11 = np.array(mass11)
mass = np.concatenate([m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11])
#m2 = pd.DataFrame(list(mass2))
#pd.concat(m1,m2)
mixed_data2 = mass.reshape(5500,2000)
```

cheking the graph

```python
plt.plot(mixed_data2[500])
plt.show()
```

save the data in excel format

```python
final_mass = pd.DataFrame(mixed_data2)
final_mass.to_csv('210809_mass_freq_with_noise_2000point_5000kDa.csv')
```
